```python
import pandas as pd
import numpy as np
import scipy.stats as stats
df = pd.read_csv('customer_behavior (2).csv')
df.head()
```

```python
mean_val = df['PurchaseAmount'].mean()
median_val = df['PurchaseAmount'].median()
mode_val = df['PurchaseAmount'].mode()[0]

print("Mean:", mean_val)
print("Median:", median_val)
print("Mode:", mode_val)
```

```python
Q1 = df['PurchaseAmount'].quantile(0.25)
Q3 = df['PurchaseAmount'].quantile(0.75)
IQR = Q3 - Q1

lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

outliers = df[(df['PurchaseAmount'] < lower) | (df['PurchaseAmount'] > upper)]

print("Number of outliers:", len(outliers))
```

```python
skewness = df['PurchaseAmount'].skew()
kurtosis = df['PurchaseAmount'].kurt()

print("Skewness:", skewness)
print("Kurtosis:", kurtosis)
```

```python
male = df[df['Gender'] == 'Male']['PurchaseAmount']
female = df[df['Gender'] == 'Female']['PurchaseAmount']

t_stat, p_val = stats.ttest_ind(male, female)

print("p-value:", p_val)
```

```python
contingency = pd.crosstab(df['ProductCategory'], df['Churn'])

chi2, p, dof, expected = stats.chi2_contingency(contingency)

print("p-value:", p)
```

```python
groups = [group['PurchaseAmount'].values for name, group in df.groupby('Region')]

f_stat, p_val = stats.f_oneway(*groups)

print("p-value:", p_val)
```

```python
campaign_avg = df.groupby('CampaignGroup')['PurchaseAmount'].mean()
print(campaign_avg)
```

```python
stat, p_val = stats.shapiro(df['PurchaseAmount'])

print("p-value:", p_val)
```

```python
sample_means = []

for i in range(1000):
    sample = df['PurchaseAmount'].sample(30)
    sample_means.append(sample.mean())

print("Mean of sample means:", np.mean(sample_means))
```

```python
mean = df['PurchaseAmount'].mean()
std = df['PurchaseAmount'].std()
n = len(df)

margin = 1.96 * (std / np.sqrt(n))

lower = mean - margin
upper = mean + margin

print("95% Confidence Interval:", (lower, upper))
```
