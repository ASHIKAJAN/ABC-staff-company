```python
import pandas as pd
import numpy as np
import scipy.stats as stats
df = pd.read_csv('housing.csv')
df.head()
```

Loading and Preprocessing


```python
from sklearn.datasets import fetch_california_housing
import pandas as pd
from sklearn.preprocessing import StandardScaler

data = fetch_california_housing()

df = pd.DataFrame(data.data, columns=data.feature_names)
df['Price'] = data.target

df.fillna(df.mean(), inplace=True)

X = df.drop('Price', axis=1)
y = df['Price']

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

X_scaled_df = pd.DataFrame(X_scaled, columns=X.columns)
```

Regression Algorithm Implementation

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from sklearn.svm import SVR

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

# Linear Regression
lr = LinearRegression()
lr.fit(X_train, y_train)
y_pred_lr = lr.predict(X_test)

# Decision Tree Regressor
dt = DecisionTreeRegressor(random_state=42)
dt.fit(X_train, y_train)
y_pred_dt = dt.predict(X_test)

# Random Forest Regressor
rf = RandomForestRegressor(random_state=42)
rf.fit(X_train, y_train)
y_pred_rf = rf.predict(X_test)

# Gradient Boosting Regressor
gb = GradientBoostingRegressor(random_state=42)
gb.fit(X_train, y_train)
y_pred_gb = gb.predict(X_test)

# Support Vector Regressor
svr = SVR()
svr.fit(X_train, y_train)
y_pred_svr = svr.predict(X_test)
```

**Linear Regression**
Works by fitting a straight line (or hyperplane) that minimizes the difference between predicted and actual values using least squares.
Suitable because it provides a simple baseline and helps understand linear relationships between housing features and prices.

---

**Decision Tree Regressor**
Splits the dataset into smaller subsets based on feature values, creating a tree where each leaf represents a predicted value.
Suitable because it can capture non-linear relationships and interactions between features without requiring scaling.

---

**Random Forest Regressor**
An ensemble of multiple decision trees built on different subsets of data; predictions are averaged to improve accuracy and reduce overfitting.
Suitable because it handles complex patterns well and is robust to noise in the housing dataset.

---

**Gradient Boosting Regressor**
Builds trees sequentially, where each new tree corrects the errors of the previous ones. Focuses on minimizing prediction errors step by step.
Suitable because it often achieves high accuracy on structured/tabular data like housing prices.

---

**Support Vector Regressor (SVR)**
Finds a function that fits the data within a certain margin (tolerance), using kernel functions to handle non-linear relationships.
Suitable because it works well with high-dimensional data and can model non-linear patterns after scaling features.


```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

models = {
    "Linear Regression": y_pred_lr,
    "Decision Tree": y_pred_dt,
    "Random Forest": y_pred_rf,
    "Gradient Boosting": y_pred_gb,
    "SVR": y_pred_svr
}

results = {}

for name, y_pred in models.items():
    mse = mean_squared_error(y_test, y_pred)
    mae = mean_absolute_error(y_test, y_pred)
    r2 = r2_score(y_test, y_pred)

    results[name] = {"MSE": mse, "MAE": mae, "R2": r2}

# Display results
for model, metrics in results.items():
    print(f"{model}:")
    print(f"  MSE: {metrics['MSE']}")
    print(f"  MAE: {metrics['MAE']}")
    print(f"  R2: {metrics['R2']}")
    print()

# Identify best and worst model based on R2 score
best_model = max(results, key=lambda x: results[x]['R2'])
worst_model = min(results, key=lambda x: results[x]['R2'])

print("Best Model:", best_model)
print("Worst Model:", worst_model)
```
