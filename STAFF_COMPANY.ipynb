```python
import pandas as pd
df=pd.read_csv("ABC Company dataset.csv")
df["Height"] = np.random.randint(150, 181, len(df))
df
```

```python
#Discover which team and position have the highest salary expenditure.
t = df.groupby("Team")["Salary"].sum()
p = df.groupby("Position")["Salary"].sum()
print("HST:", t.idxmax())
print("HSP:", p.idxmax())
```

```python
t.plot(kind="bar")
plt.title("Salary by Team")
plt.ylabel("Total Salary")
plt.show()
p.plot(kind="bar")
plt.title("Salary by Position")
plt.ylabel("Total Salary")
plt.show()
```

SALARY BY TEAM AND POSITION

The position with the highest bar receives the largest total salary.

This helps understand which roles are most financially significant in the company.

```python
#Investigate if there's any correlation between age and salary, and represent it visually.
import matplotlib.pyplot as plt
c = df["Age"].corr(df["Salary"])
print("Correlation:", round(c, 2))
plt.scatter(df["Age"], df["Salary"])
plt.xlabel("Age")
plt.ylabel("Salary")
plt.title("Age vs Salary")
plt.show()
```

AGE AND SALARY
The correlation analysis and scatter plot show the relationship between employees' age and salary. The value indicates whether salary increases, decreases, or has no strong relation with age."
